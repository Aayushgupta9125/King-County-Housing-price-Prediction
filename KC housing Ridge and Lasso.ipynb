```python
import pandas as pd 
data=pd.read_csv("kc_house_data.csv")
data.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error


# Prepare the data
# Remove unnecessary columns
data = data.drop(columns=['id', 'date'])

# Split data into training and testing sets
train_data, test_data = train_test_split(data, test_size=0.2, random_state=123)

# Separate features and target variable
train_x = train_data.drop(columns='price')
train_y = train_data['price']
test_x = test_data.drop(columns='price')
test_y = test_data['price']

# Ridge regression
ridge_model = Ridge()
ridge_model.fit(train_x, train_y)
ridge_pred = ridge_model.predict(test_x)
ridge_mse = mean_squared_error(test_y, ridge_pred)

# Lasso regression
lasso_model = Lasso()
lasso_model.fit(train_x, train_y)
lasso_pred = lasso_model.predict(test_x)
lasso_mse = mean_squared_error(test_y, lasso_pred)

print(f"Ridge Regression MSE: {ridge_mse}")
print(f"Lasso Regression MSE: {lasso_mse}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score


# Prepare the data
# Remove unnecessary columns
data = data.drop(columns=['id', 'date'])

# Split data into training and testing sets
train_data, test_data = train_test_split(data, test_size=0.2, random_state=123)

# Separate features and target variable
train_x = train_data.drop(columns='price')
train_y = train_data['price']
test_x = test_data.drop(columns='price')
test_y = test_data['price']

# Ridge regression
ridge_model = Ridge()
ridge_model.fit(train_x, train_y)
ridge_pred = ridge_model.predict(test_x)
ridge_mse = mean_squared_error(test_y, ridge_pred)
ridge_r2 = r2_score(test_y, ridge_pred)

# Lasso regression
lasso_model = Lasso()
lasso_model.fit(train_x, train_y)
lasso_pred = lasso_model.predict(test_x)
lasso_mse = mean_squared_error(test_y, lasso_pred)
lasso_r2 = r2_score(test_y, lasso_pred)

print(f"Ridge Regression MSE: {ridge_mse}")
print(f"Ridge Regression R²: {ridge_r2}")
print(f"Lasso Regression MSE: {lasso_mse}")
print(f"Lasso Regression R²: {lasso_r2}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Distribution of numerical features
numerical_features = ['sqft_living', 'sqft_lot', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'sqft_living15', 'sqft_lot15']
data[numerical_features].hist(bins=30, figsize=(15, 10))
plt.suptitle('Distribution of Numerical Features')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(8, 4))
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Pairplot of some features
sample_features = ['price', 'sqft_living', 'bedrooms', 'bathrooms', 'floors', 'grade']
sns.pairplot(data[sample_features])
plt.suptitle('Pairplot of Selected Features')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



# Bar plots to see the relationship between categorical variables and price

# Price vs. Number of Bedrooms
bedrooms_avg_price = data.groupby('bedrooms')['price'].mean().reset_index()
plt.figure(figsize=(11, 4))
sns.barplot(x='bedrooms', y='price', data=bedrooms_avg_price)
plt.title('Average Price vs. Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Average Price')
plt.show()

# Price vs. Number of Bathrooms
bathrooms_avg_price = data.groupby('bathrooms')['price'].mean().reset_index()
plt.figure(figsize=(11, 4))
sns.barplot(x='bathrooms', y='price', data=bathrooms_avg_price)
plt.title('Average Price vs. Number of Bathrooms')
plt.xlabel('Number of Bathrooms')
plt.ylabel('Average Price')
plt.show()

# Price vs. Number of Floors
floors_avg_price = data.groupby('floors')['price'].mean().reset_index()
plt.figure(figsize=(11, 4))
sns.barplot(x='floors', y='price', data=floors_avg_price)
plt.title('Average Price vs. Number of Floors')
plt.xlabel('Number of Floors')
plt.ylabel('Average Price')
plt.show()
```

```python
# Scatter plots to see relationships between numerical variables and price
plt.figure(figsize=(8, 4))
sns.scatterplot(x='sqft_living', y='price', data=data)
plt.title('Price vs. Square Footage of Living Area')
plt.xlabel('Square Footage of Living Area')
plt.ylabel('Price')
plt.show()

plt.figure(figsize=(8, 4))
sns.scatterplot(x='sqft_lot', y='price', data=data)
plt.title('Price vs. Square Footage of Lot')
plt.xlabel('Square Footage of Lot')
plt.ylabel('Price')
plt.show()

plt.figure(figsize=(8, 4))
sns.scatterplot(x='grade', y='price', data=data)
plt.title('Price vs. Grade')
plt.xlabel('Grade')
plt.ylabel('Price')
plt.show()
```

```python
# Geographical distribution of house prices
plt.figure(figsize=(8, 4))
sns.scatterplot(x='long', y='lat', hue='price', palette='viridis', data=data)
plt.title('Geographical Distribution of House Prices')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
